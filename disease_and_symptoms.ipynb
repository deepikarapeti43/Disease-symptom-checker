```python
# Importing Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Create a SQLAlchemy engine
engine = create_engine('sqlite:///DiseasesAndSymptoms.db')

# Query the database using SQLAlchemy
query = "SELECT * FROM DiseasesAndSymptoms;" 
disease_df = pd.read_sql(query, engine)
```

```python
# Load CSV for disease data. 
#disease_df = pd.read_csv('./disease_and_symptoms_data/DiseaseAndSymptoms.csv')
disease_df.head()
```

```python
# Determine the number of null values in each symptom column
disease_df.isna().sum()
```

```python
# Determine the number of unique values in each column.
unique_counts = disease_df.nunique()
unique_counts
```

```python
# Split our preprocessed data into our features and target arrays
y = disease_df.Disease.values.reshape(-1, 1)
X = disease_df.drop("Disease", axis=1)
feature_names = X.columns
```

```python
# Review the y variable Series
y[:5]
```

```python
# Review the X variable DataFrame
X.head()
```

```python
label_encoder = LabelEncoder()

# Iterate over each column in the DataFrame
for column in X.columns:
    if X[column].dtype == 'object':  # Check if the column contains string values
        # Replace NaN values with a placeholder, for example, 'Unknown'
        X[column].fillna('Unknown', inplace=True)
        # Fit label encoder and transform the labels in the column
        X[column] = label_encoder.fit_transform(X[column])

y = label_encoder.fit_transform(y)
```

```python
y
```

```python
X.isna().sum()
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
X.describe()
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Using random forest model
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=50, random_state=78)
rf_model
```

```python
#Fitting model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
#Making predictions using test data
predictions = rf_model.predict(X_test_scaled)
```

```python
#Accuracy score
acc_score = accuracy_score(y_test, predictions)
acc_score
```
